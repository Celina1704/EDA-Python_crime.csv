### 5. On what days and during which hours are the most crimes committed?
>Monday, Tuesday, Saturday, Sunday @17

```python
week_and_hour = df.groupby(['HOUR', 'DAY_OF_WEEK']).count()['INCIDENT_NUMBER'].unstack() #unstack, hour = row, day = col
print(week_and_hour)
```

```python
week_and_hour.columns = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
sns.heatmap(week_and_hour, cmap=sns.cubehelix_palette(as_cmap=True))
#the darker means more crime
```

### 6. In which months were the number of crimes below average?
### 7. In which months on average did the most crimes occur?

```python
#if the values is less than the average crime per month, highlight the value in blue

#counts the occurrences of 'INCIDENT_NUMBER' for each group.
val = df.groupby(['YEAR', 'MONTH']).count()['INCIDENT_NUMBER']
avg_crime = val.mean()
print("The average number of crimes is " + str(avg_crime))

year_and_month = val.unstack()

#if the values is less than the average crime per month, highlight the value in blue
def style_negative(v, props=''):
  return props if v < avg_crime else None

S2 = year_and_month.style.applymap(style_negative, props='color:blue;')\

S2

#v = value of a cell in the DataFrame
#props = style properties
#applymap =  function applies a function to each element of the DataFrame
```

```python
#use apply function to highlight the maximum in a column in darkgreen

def highlight_max(s, props=''):
  return np.where(s == np.nanmax(s.values), props, '')

S2.apply(highlight_max, props='color:white;background-color:darkgreen', axis=0)
```

### 8. In which districts were the most crimes commmitted on yearly basis?

> All year B2

```python
district_and_year = df.groupby(['YEAR', 'DISTRICT'])['INCIDENT_NUMBER'].count().unstack()
print(district_and_year)
```

```python
plt.figure(figsize=(40,24))
district_and_year.plot(kind='bar', stacked=True)
```

```python
sns.heatmap(district_and_year, cmap=sns.cubehelix_palette(as_cmap=True))
```

```python
val = df.groupby(['YEAR', 'DISTRICT']).count()['INCIDENT_NUMBER']
avg_crimes_per_district_and_year = val.mean()
print("The average number of crimes based on districts and year is " + str(avg_crimes_per_district_and_year))

year_and_district = val.unstack()

year_and_district
```

```python
districts_with_max_crimes = year_and_district.idxmax(axis=1)
print(districts_with_max_crimes)

#idxmax = returns the index of the first occurrence of the maximum value along a specified axis
```

```python
def style_most(v, props=''):
  return props if v < avg_crimes_per_district_and_year else None

# S3 = year_and_district.style.applymap(style_most, props='color:white;background-color:purple')
S3 = year_and_district.style.applymap(style_most, props='color:thistle')


def highlight_max2(s, props=''):
  return np.where(s == np.nanmax(s.values), props, '')

S3.apply(highlight_max2, props='color:blue;background-color:#8b795e;font-weight:bold', axis=0)

#np.nanmax() = returns the maximum value of an array, ignoring any NaN values
```
